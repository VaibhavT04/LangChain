```python
from langchain_community.document_loaders import WebBaseLoader
import bs4
loaderr=WebBaseLoader(web_paths=("https://en.wikipedia.org/wiki/UEFA_Champions_League",), 
                      bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                          class_=("mw-page-title-main", "hatnote navigation-not-searchable")
                      )))
loaderr.load()
```
